```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/checked.csv')

df.head()
```

```python
df['Date'] = pd.to_datetime(df['fecha_servidor']) 
df.head()
```

```python
df.drop(columns=['fecha_servidor', 'fecha_esp32', 'MAC'], inplace=True)
```

```python
df.head()
```

```python
df = df.rename(columns={'Date': 'timestamp'})
```

```python
df.head()
```

```python
df.set_index('timestamp', inplace=True)
```

```python
df_downsampled = df.resample('1T').mean().round(5)
df_downsampled.head()
```

```python
df_downsampled.fillna(method='ffill', inplace=True)  # Forward fill
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df_downsampled['CPU temperature - Centigrade Degrees (°C)'], label='Temperature (Downsampled)', color='purple')
plt.title('Downsampled Temperature Data')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(range(len(df_downsampled)), df_downsampled['CPU temperature - Centigrade Degrees (°C)'].values, color='purple', label='Temperature', s=1)
plt.title('Server Temperature Points')
plt.xlabel('Index')
plt.ylabel('Server Temperature (°C)')
plt.legend()
plt.show()
```

```python
df = df_downsampled[~((df_downsampled.index >= '2021-09-19 21:15:00') & 
          (df_downsampled.index <= '2021-11-24 15:51:00') & 
          (df_downsampled['CPU temperature - Centigrade Degrees (°C)'] == 26.41379))]
```

```python
temp_counts = df['CPU temperature - Centigrade Degrees (°C)'].value_counts()
top_n = 20  

top_temp_counts = temp_counts.head(top_n)

plt.figure(figsize=(10, 6))
top_temp_counts.plot(kind='bar')
plt.title(f'Top {top_n} Most Frequent CPU Temperatures')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
